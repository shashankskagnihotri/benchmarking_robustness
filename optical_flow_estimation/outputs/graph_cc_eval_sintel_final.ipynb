```python
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folders = ['ccmr_sintel','craft_sintel','dicl_sintel','dip_sintel','fastflownet_sintel','flow1d_sintel','flowformer_sintel','flowformer++_sintel','gma_sintel','gmflow_sintel','hd3_sintel','irr_pwc_sintel','liteflownet2_sintel','liteflownet3_sintel','llaflow_sintel','maskflownet_sintel','maskflownets_sintel','ms-raft+_mixed','raft_sintel','rapidflow_sintel','rpknet_sintel','scopeflow_sintel','skflow_sintel','starflow_sintel','videoflow_bof_sintel' ]  # 这里可以添加更多文件夹名称

json_file_paths = {folder: os.path.join('validate', folder, 'metrics_sintel-final.json') for folder in folders}


def process_json(json_file_path):

    with open(json_file_path, 'r') as file:
        data = json.load(file)

    # Extracting data
    results = []
    for experiment in data['experiments']:
        if experiment['attack'] == 'common_corruptions':
            attack_name = experiment['name']
            epe = experiment['metrics']['epe']
            results.append([attack_name, epe])

    # Convert to DataFrame
    df = pd.DataFrame(results, columns=['corruption_name', 'epe'])
    return df

# Process each JSON file and save to a DataFrame
dfs = {key: process_json(path) for key, path in json_file_paths.items()}

# Saving to different pages of a CSV file
with pd.ExcelWriter('/pfs/work7/workspace/scratch/ma_xinygao-team_project_fss2024/benchmarking_robustness/optical_flow_estimation/outputs/tables/cc_sintel_final_metrics_summary.xlsx', engine='xlsxwriter') as writer:
    for i, (key, df) in enumerate(dfs.items(), start=1):
        # Save to different worksheets, the worksheet name is the folder name, and the index starts at 1
        sheet_name = key
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```

```python
excel_file_path = '/pfs/work7/workspace/scratch/ma_xinygao-team_project_fss2024/benchmarking_robustness/optical_flow_estimation/outputs/tables/cc_sintel_final_metrics_summary.xlsx'

excel_file = pd.ExcelFile(excel_file_path)

sheet_names = excel_file.sheet_names

# Initialize an empty list to store the epe summary results of each model
summary_results = []

# Iterate through each worksheet and sum up epe
for sheet in sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet)
    
    # Summarize (sum) the epes for the same corruption_name
    df_summary = df.groupby('corruption_name')['epe'].sum().reset_index()
    
    # Add a new column to store the model name (worksheet name)
    df_summary['model'] = sheet

    summary_results.append(df_summary)

# Combine the summary results of all models into one DataFrame
final_df = pd.concat(summary_results, ignore_index=True)

final_df = final_df[['model', 'corruption_name', 'epe']]

final_df['model'] = final_df['model'].str.split('_').str[0]

# Classification by corruption_name
grouped_by_corruption = final_df.groupby('corruption_name')

# The directory path where the images are saved
output_dir = '/pfs/work7/workspace/scratch/ma_xinygao-team_project_fss2024/benchmarking_robustness/optical_flow_estimation/outputs/graphs/common_corruptions/sintel_final'  # 替换为你想保存图像的路径

# If the directory does not exist, create it
os.makedirs(output_dir, exist_ok=True)

for corruption_name, df in grouped_by_corruption:
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x='model', y='epe', data=df, palette='viridis')
    plt.title(f'EPE by Model for {corruption_name}')
    plt.xlabel('Model')
    plt.ylabel('EPE')
    plt.xticks(rotation=45)  
    plt.tight_layout()  # Adjust layout to prevent labels from overlapping
    
    # Label the epe value at the top of each bar
    for i in range(len(df)):
        value = round(df['epe'].iloc[i], 2)  # Keep 2 decimal places
        ax.text(i, df['epe'].iloc[i] + 0.1, str(value), ha='center', va='bottom')
    
    save_path = os.path.join(output_dir, f'cc_{corruption_name}_epe_bar_sintel_final.png')
    
    plt.savefig(save_path)

    plt.show()
```
